## feature scaling
1) 표준화 standardization
2) outlier 제거 ZSCORE 
3) 정규화 normalization 

```python
from sklearn.datasets import load_iris
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import scipy as sp
import scipy.stats
from sklearn.preprocessing import StandardScaler
```

```python
iris = load_iris()
iris_data = iris.data
iris_df = pd.DataFrame(data = iris_data, columns = iris.feature_names)
```

### 표준화

### scaling process

(1) training data -> fit -> 객체에 저장

(2) training data -> transform

(3) test data 는  trasform 만 진행→ fit 된 분포를 그대로 적용

```python
#mean
print(iris_df.mean())
#var
print(iris_df.var())

# https://m.blog.naver.com/PostView.naver?isHttpsRedirect=true&blogId=demian7607&logNo=222009975984 sklearn scaler 차이
# standard scaler 평균 0 , 분산 1로 조정합니다

scaler = StandardScaler()

#https://m.blog.naver.com/PostView.naver?isHttpsRedirect=true&blogId=kiakass&logNo=222085098701

#fit 평균 𝜇과  표준편차 𝜎를 계산
scaler.fit(iris_df)
#정규화/표준화, Standardization, z = (𝑥-𝜇)/𝜎
iris_scaled = scaler.transform(iris_df)

iris_df_scaled = pd.DataFrame(data=iris_scaled, columns=iris.feature_names)
print(iris_df_scaled.mean())
print(iris_df_scaled.var())
```

```python
#정규 분포 모양 비교
fig, ax = plt.subplots(1,2,figsize = (8,5))

print(ax[0])
print(ax[1])

x0 = iris_df['petal length (cm)'].dropna().values #NaN values drop
sns.distplot(x0, kde=False, rug=False, fit = sp.stats.norm, ax = ax[0]) #커널 밀도 함수, 러그밀도함수 는 안보이고(False) / 정규분포만 보이게(Fit)
#정규분포 그래프는 scipy의 stats의 norm으로 확인 할 수 있다. (sp.stats.norm)

x1 = iris_df_scaled['petal length (cm)'].values
sns.distplot(x0, kde=False, rug=False, fit = sp.stats.norm, ax = ax[1])
plt.show()



fig, ax = plt.subplots(1,2,figsize = (8,5))

x2 = iris_df['sepal length (cm)'].dropna().values #NaN values drop
sns.distplot(x2, kde=False, rug=False, fit = sp.stats.norm, ax = ax[0]) #커널 밀도 함수, 러그밀도함수 는 안보이고(False) / 정규분포만 보이게(Fit)
#정규분포 그래프는 scipy의 stats의 norm으로 확인 할 수 있다. (sp.stats.norm)


x3 = iris_df_scaled['sepal length (cm)'].values
sns.distplot(x3, kde=False, rug=False, fit = sp.stats.norm, ax = ax[1])
plt.show()
```

### outlier 제거

```python
iris_df_scaled
```

```python
df_Zscore = pd.DataFrame()
outlier_dict = {}
outlier_idx_list = []

for one_col in iris_df_scaled.columns:
    print("check",one_col)
    df_Zscore[f'{one_col}_Zscore'] = sp.stats.zscore(iris_df_scaled[one_col])
    outlier_dict[one_col] = df_Zscore[f'{one_col}_Zscore'][(df_Zscore[f'{one_col}_Zscore']>2)|
                                                          (df_Zscore[f'{one_col}_Zscore']<-2)]
    outlier_idx_list.append(list(outlier_dict[one_col].index))
    if len(outlier_dict[one_col]):
        print(one_col, 'Has outlier\n', outlier_dict[one_col])
    else:
        print(one_col, "Has Not outlier")
    print()
    
print("Before", iris_df_scaled.shape)
all_outlier_idx = sum(outlier_idx_list,[])
iris_df_scaled = iris_df_scaled.drop(all_outlier_idx)
print("After (drop outlier)", iris_df_scaled.shape)
```

### 정규화
